```python
%matplotlib inline
import mxnet as mx
import numpy as np
import cv2
import matplotlib.pyplot as plt
import logging
import sys, datetime, math, random

logger = logging.getLogger()
logger.setLevel(logging.DEBUG)
```

```python
data = mx.symbol.Variable('data')
fc1 = mx.symbol.FullyConnected(data = data, name = 'fc1', num_hidden=128)
act1 = mx.symbol.Activation(data = fc1, name = 'relu1', act_type='relu')
fc2 = mx.symbol.FullyConnected(data = act1, name = 'fc2', num_hidden = 64)
act2 = mx.symbol.Activation(data = fc2, name = 'relu2', act_type = 'relu')
fc3 = mx.symbol.FullyConnected(data = act2, name = 'fc3', num_hidden = 10)
mlp = mx.symbol.SoftmaxOutput(data = fc3, name = 'softmax')
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
np.random.seed(1234)
p = np.random.permutation(mnist.data.shape[0])
X = mnist.data[p]
Y = mnist.target[p]

for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(X[i].reshape((28,28)), cmap='Greys_r')
    plt.axis('off')
plt.show()
```

```python
X = X.astype(np.float32)/255
X_train = X[:60000]
Y_train = Y[:60000]
X_test = X[60000:]
Y_test = Y[60000:]
```

```python
batch_size = 100
train_iter = mx.io.NDArrayIter(X_train, Y_train, batch_size = batch_size)
test_iter  = mx.io.NDArrayIter(X_test,  Y_test,  batch_size = batch_size)
```

使用新的model.Module()+model.fit()参数

```python
model = mx.mod.Module(
    symbol = mlp,
    context = mx.gpu(0),
    data_names = ['data'],
    label_names = ['softmax_label']
)
model.fit(
    train_iter,
    eval_data = test_iter,
    optimizer = 'sgd',
    num_epoch = 10,
    optimizer_params = {'learning_rate':0.1, 'momentum':0.9, 'wd':0.00001},
    eval_metric = 'acc',
    batch_end_callback = mx.callback.Speedometer(batch_size, 200)
)
```

使用旧的model.feedback()+model.fit()参数

```python
model = mx.model.FeedForward(
    ctx = mx.gpu(0),      # Run on GPU 0
    symbol = mlp,         # Use the network we just defined
    num_epoch = 10,       # Train for 10 epochs
    learning_rate = 0.1,  # Learning rate
    momentum = 0.9,       # Momentum for SGD with momentum
    wd = 0.00001)         # Weight decay for regularization
model.fit(
    X=train_iter,  # Training data set
    eval_data=test_iter,  # Testing data set. MXNet computes scores on test set every epoch
    batch_end_callback = mx.callback.Speedometer(batch_size, 200))
```

使用debug模式，给model.fit传入monitor参数，具体官方文档解释为：
DNNs can perform poorly for a lot of reasons, like learning rate too big/small, initialization too big/small, network structure not reasonable, etc. When this happens it's often helpful to print out the weights and intermediate outputs to understand what's going on. MXNet provides a monitor utility that does this:

```python
def norm_stat(d):
    return mx.nd.norm(d)/np.sqrt(d.size)
mon = mx.mon.Monitor(
    100,
    norm_stat,
    pattern = '.*weight',
    sort = True
)
model = mx.mod.Module(
    symbol = mlp,
    context = mx.gpu(0),
    data_names = ['data'],
    label_names = ['softmax_label']
)
model.fit(
    train_iter,
    eval_data = test_iter,
    optimizer = 'sgd',
    num_epoch = 10,
    optimizer_params = {'learning_rate':0.1, 'momentum':0.9, 'wd':0.00001},
    eval_metric = 'acc',
    batch_end_callback = mx.callback.Speedometer(batch_size, 200),
    monitor = mon
)
```

定义自己的神经网络training loop，官方文档解释如下：
mx.model.Module is a convenience wrapper for training standard feed forward networks. What if the model you are working with is more complicated? With MXNet, you can easily control every aspect of training by writing your own training loop.
Neural network training typically has 3 steps: forward, backward (gradient), and update. With custom training loop, you can control the details in each step as while as insert complicated computations in between. You can also connect multiple networks together.

```python
model = mx.mod.Module(
    symbol = mlp,
    context = mx.gpu(0),
    data_names = ['data'],
    label_names = ['softmax_label']
)
```

```python
model.bind(data_shapes  = train_iter.provide_data,
           label_shapes = train_iter.provide_label)
model.init_params()
```

```python
opt = mx.optimizer.SGD(
    learning_rate = 0.1,
    momentum = 0.9,
    wd = 0.00001,
    rescale_grad = 1.0/train_iter.batch_size
)
updater = mx.optimizer.get_updater(opt)
```

```python
metric = mx.metric.Accuracy()
```

```python
for epoch in range(10):
    train_iter.reset()
    metric.reset()
    t = 0
    for batch in train_iter:
        data[:] = batch.data[0]
        label[:] = batch.label[0]
        
        model.forward(batch_size)
        model.backward()
        model.update()
        metric.update()
        
        t += 1
        if t % 100 == 0:
            print 'epoch:', epoch, 'iter:', t, 'metric:', metric.get()
```
